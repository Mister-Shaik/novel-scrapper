```python
from bs4 import BeautifulSoup as bs, SoupStrainer
```

```python
import requests
page = requests.get("https://boxnovel.com/novel/i-alone-level-up/");
```

```python
soup = bs(page.content, 'html.parser')
```

```python
ulit = soup.find_all('li', class_='wp-manga-chapter')
arr = []
```

```python
for x in ulit:
    arr.extend([y.get_attribute_list('href') for y in x.find_all('a')][0])
```

```python
arr.reverse()
tot = []
```

```python
for c in arr:
    chap = requests.get(c)
    chapsoup = bs(chap.content, 'html.parser')
    chawords = chapsoup.find_all('p')
    tot.append(chawords)
```
